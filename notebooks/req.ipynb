```python
import requests
from typing import Tuple, List, Any
from time import sleep
import pickle
```

```python
def obter_municipio_sp() -> List[Tuple[Any, Any]]:
    url = 'https://servicodados.ibge.gov.br/api/v1/localidades/estados/35/municipios?view=nivelado'
    req = requests.get(url=url)
    req = req.json()
    municipios = [(mun['municipio-id'], mun['municipio-nome']) for mun in req]
    return municipios
```

```python
mun_sp = obter_municipio_sp()
```

```python
mun_sp
```

```python
def guardar_objeto(objeto):
    with open('../municipios_sp/municipios.pkl', 'wb') as arquivo:
        pickle.dump(objeto, arquivo)
```

```python
guardar_objeto(mun_sp)
```

```python
def carregar_objeto():
    with open('../municipios_sp/municipios.pkl', 'rb') as arquivo:
        lista_mun = pickle.load(arquivo)
    return lista_mun
```
